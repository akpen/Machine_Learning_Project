```python
import cv2
vidcap = cv2.VideoCapture('data/video1/cars.mp4')
success,image = vidcap.read()
count = 1000
success = True
while success:
    success,image = vidcap.read()
    print('Read a new frame: ', success)
    image_compress = cv2.resize(image,None,fx=0.2, fy=0.2, interpolation = cv2.INTER_CUBIC)    #Compressing Image
    #image_compress = image[20:70,5:105]
    cv2.imwrite("data/video1/images1/frame%d.jpg" % count, image)     # save frame as JPEG file
    cv2.imwrite("data/video1/images1_compress/frame_compress%d.jpg" % count, image_compress)     # save frame as JPEG file
    count += 1
```
